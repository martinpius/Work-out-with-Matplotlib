```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
from random import randint
%matplotlib inline
```

```python
#Creating fake data.
random.seed(2)
dev_x = []
for i in range(100):
    dev = randint(22,35)
    dev_x.append(dev)
    
y_py = []
for i in range(100):
    ypy = randint(36000,88000)
    y_py.append(ypy)
    
y_dev =[]
for i in range(100):
    ydev = randint(32000,85000)
    y_dev.append(ydev)

y_js = []
for i in range(100):
    yjs = randint(33500,87900)
    y_js.append(yjs)
    
```

```python
#Re-arrange the data in ascending order
dev_x = sorted(dev_x)
y_py = sorted(y_py)
y_dev = sorted(y_dev)
y_js = sorted(y_js)

```

```python
#Line plot for the data.
plt.style.use('fivethirtyeight')
plt.plot(dev_x, y_py,label = 'Python',linewidth = 3, color = 'mediumvioletred')
plt.plot(dev_x, y_dev,label = 'Others',linewidth = 3, color = 'indigo')
plt.plot(dev_x, y_js, label = 'Java Scripts', linewidth = 3, color = 'magenta')
plt.title("Developer's salary by age")
plt.xlabel('Age in (years)')
plt.ylabel('Salary in (USD)')
plt.legend(loc = 'upper left')
plt.tight_layout()
```

```python
#The bar plot for the data
#Creating fake data.
random.seed(2)
dev_x = []
for i in range(11):
    dev = randint(22,35)
    dev_x.append(dev)
    
y_py = []
for i in range(11):
    ypy = randint(36000,88000)
    y_py.append(ypy)
    
y_dev =[]
for i in range(11):
    ydev = randint(32000,85000)
    y_dev.append(ydev)

y_js = []
for i in range(11):
    yjs = randint(33500,87900)
    y_js.append(yjs)
```

```python
#Ordering and change the x's into numpy arrays
dev_x = sorted(dev_x)
index_x = np.arange(len(dev_x))
width = 0.3# to shift the bars side by side
y_py = sorted(y_py)
y_dev = sorted(y_dev)
y_js = sorted(y_js)
```

```python
plt.style.use('fivethirtyeight')
plt.bar(index_x -width, y_py, label = 'Python', color = 'indigo', width = width)
plt.bar(index_x, y_dev, label = 'Others', color = 'mediumvioletred', width = width)
plt.bar(index_x+width, y_js, label = 'Java Scripts', color = 'magenta', width = width)
plt.title("Developer's salary per age")
plt.xlabel('Age in (years)')
plt.ylabel('Salary in (USD)')
plt.legend(loc = 'upper left',bbox_to_anchor = (1.5,1.3))
plt.tight_layout()
```

```python
#Real data application
data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/02-BarCharts/data.csv')
```

```python
data.head(5)
```

```python
#Create a new language column with list of strings as language
```

```python
data['Language'] = data.loc[:,'LanguagesWorkedWith'].str.split(';')
```

```python
data.Language.head(3)
```

```python
#Iterate through to obtain the frequency for each language
lang_counter = Counter()
for row in data.Language:
    lang_counter.update(row)
```

```python
lang_counter
```

```python
#Obtain the top 20 most popular Programming languages
top_20_lang = lang_counter.most_common(20)
```

```python
top_20_lang
```

```python
#Extract the axes from the list of tupples
languages = []
freqs = []
for lans in top_20_lang:
    languages.append(lans[0])
    freqs.append(lans[1])
```

```python
languages
```

```python
freqs
```

```python
#Obtain the plot for the top-20 Most popular Programming language
languages.reverse()
freqs.reverse()
plt.style.use('fivethirtyeight')
plt.barh(languages, freqs, color = 'magenta')
plt.title('Top 20 Most Popular Programing Languages')
plt.xlabel('Number of prgrammers')
plt.ylabel('Programming language')
plt.tight_layout()
```
